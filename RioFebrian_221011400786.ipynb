````python
# This cell defines the employee_detail_list variable.
# Make sure to execute this cell first before running the next one.
employee_detail_list = [
     {
          'name': 'EMP-0001',
          'first_name': 'John',
          'last_name': 'Doe',
          'full_name': 'John Doe',
          'company': 'Company 1',
          'nationality': 'Australia'
      },
      {
          'name': 'EMP-0002',
          'first_name': 'Tom',
          'last_name': 'Smith',
          'full_name': 'Tom Smith',
          'company': 'Company 2',
          'nationality': 'United States'
      },
      {
          'name': 'EMP-0003',
          'first_name': 'Andrew',
          'last_name': 'Sebastian',
          'full_name': 'Andrew Sebastian',
          'company': 'Company 3',
          'nationality': 'United States'
      },
      {
          'name': 'EMP-0005',
          'first_name': 'Ying Han',
          'last_name': 'Tan',
          'full_name': 'Ying Han Tan',
          'company': 'Company 1',
          'nationality': 'Australia'
      },
      {
          'name': 'EMP-0015',
          'first_name': 'Kenneth',
          'last_name': 'Ng',
          'full_name': 'Kenneth Ng',
          'company': 'Company 3',
          'nationality': 'United States'
      },
      {
          'name': 'EMP-0018',
          'first_name': 'Rubby',
          'last_name': 'Lee',
          'full_name': 'Rubby Lee',
          'company': 'Company 2',
          'nationality': 'Hong Kong'
      },
     {
          'name': 'EMP-0017',
          'first_name': 'Robert',
          'last_name': 'White',
          'full_name': 'Robert White',
          'company': 'Company 1',
          'nationality': 'United Kingdom'
      }
  ]

def list_employees_by_company_domain(company_detail_list, employee_detail_list):
    """
    List all Employees work By company domain.

    Args:
        company_detail_list (list): List of company details.
        employee_detail_list (list): List of employee details.

    Returns:
        dict: A dictionary with company domains as keys and a list of employee names as values.
    """
    employees_by_domain = {}
    # Create a dictionary mapping company name to domain
    company_domain_map = {company['name']: company['domain'] for company in company_detail_list}

    for employee in employee_detail_list:
        # Get the company domain using the company name from employee details
        company_domain = company_domain_map.get(employee['company'])

        # If the company domain is found, add the employee's full name to the list
        if company_domain:
            if company_domain not in employees_by_domain:
                employees_by_domain[company_domain] = []
            employees_by_domain[company_domain].append(employee['full_name'])

    return employees_by_domain


# Call the function to get the desired output
result = list_employees_by_company_domain(company_detail_list, employee_detail_list)
print(result)
````

````python
def get_company_employee_nationality(company_detail_list, employee_detail_list):
    """
    Mengembalikan perusahaan dengan sejumlah karyawan berkewarganegaraan yang bekerja di perusahaan tersebut.

    Args:
        company_detail_list (list): Daftar detail perusahaan.
        employee_detail_list (list): Daftar detail karyawan.

    Returns:
        list: Daftar perusahaan dengan sejumlah karyawan berkewarganegaraan.
    """

    company_nationalities = []
    for company_data in company_detail_list:
        company_name = company_data['name']
        employee_nationality = {}

        # Menghitung jumlah karyawan per kewarganegaraan untuk setiap perusahaan
        for employee_data in employee_detail_list:
            if employee_data['company'] == company_name:
                nationality = employee_data['nationality']
                employee_nationality[nationality] = employee_nationality.get(nationality, 0) + 1

        company_nationalities.append({
            "perusahaan": company_name,
            "employee_nationality": employee_nationality
        })

    return company_nationalities

# Contoh penggunaan
result = get_company_employee_nationality(company_detail_list, employee_detail_list)
print(result)
````
